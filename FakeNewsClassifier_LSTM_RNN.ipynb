```python
## importing tensorflow
import tensorflow as tf
print(tf.__version__)
```

```python
## Importing tensorflow packages

from tensorflow.keras.layers import Embedding, Dense, LSTM, Dropout
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
## vocabulary size

voc_size = 5000
```

## **Data cleaning**

```python
messages = X.copy()
messages.head()
```

```python
messages.reset_index(inplace=True)
messages.head()
```

```python
import nltk
import re
nltk.download('stopwords')
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []

for i in range(0,len(messages)):
  ## with the help of re(regular expression) library removing special characters
  ## apart from a-z and A-Z all the other alphabets or charters will be replaced by ' '
  review = re.sub('[^a-zA-Z]', ' ', str(messages['title'][i]))
  review = review.lower()
  review = review.split()
  ## applying stemming those words only which are not present in the english stopwords
  review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus
```

## **One Hot representation**

```python
## Applying one_hot encoddinfg

oneHot_repr = [one_hot(words, voc_size) for words in corpus]
oneHot_repr
```

## **Applying Padding - creating embedded_docs**

```python
## Applying padding for equalizing the number of words in a sentance
sent_length = 20
embedded_docs = pad_sequences(oneHot_repr, padding='pre', maxlen=sent_length)
# len(embedded_docs)
embedded_docs.shape
```

```python
embedded_docs[0]
```

## ***embedded_docs will be converted into final independent feature(X_final) while training the model

## **Model Creation**

```python
embedding_vector_features = 40
model = Sequential()
#### Adding Embedding layer (i/p layer)------------
model.add(Embedding(voc_size, embedding_vector_features, input_length = sent_length))

#### Adding dropout layer for increasing accuracy and reducing overfitting
model.add(Dropout(0.3))

#### Adding LSTM layer - 1 layer with 100 neurons(hidden layer)----------------------
model.add(LSTM(200))

#### Adding dropout layer for increasing accuracy and reducing overfitting
model.add(Dropout(0.3))

#### Adding Dense layer (o/p layer)-----------
model.add(Dense(1, activation='sigmoid'))

#### Compiling the model----------------
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

print(model.summary())
```

## **Final independent and dependent feature creation(X_final, y_final)**

```python
import numpy as np
X_final = np.array(embedded_docs)
y_final = np.array(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size = 0.33, random_state = 42)
```

## **Final Model Training**

```python
### model is the sequential model
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=10, batch_size=64)
```

## **Performance metrics and accuracy**







```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

## As model.predict_classess has been depricated in tensorflow 2.6, and as it is a binary classificatin, we have used the below code
## For multi class classification --->  y_pred = np.argmax(model.predict(X_test), axis=-1)
y_pred = (model.predict(X_test) > 0.5).astype("int32")
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_pred, y_test)
```
